Hallar la probabilidad de que {$x+y\leq1$} y  {$x\cdot y\leq\frac{2}{9}$}

2. Mediante simulación.

**Analiticamente:**

Dado que $X,Y$ son variables aleatorias que se distribuyen uniformemente en $(0,1)$ y, además, asumimos independencia, podemos considerar la función de densidad conjunta

\begin{equation*}
\begin{aligned}
f_{Y,Y}(x,y)=\frac{1}{1}I_{(0,1)}(x) I_{(0,1)}(y)
\end{aligned}
\end{equation*}

la cual tiene como soporte el cuadrado $[0,1]$X$[0,1]$ en el plano XY

Definimos las variables aleatorias $Z_1$, $Z_2$ como sigue
\begin{equation*}
\begin{aligned}
Z_1 = X+Y\\
Z_2 = XY
\end{aligned}
\end{equation*}

y nos preguntamos por la probabilidad

$$P(\{x+y\leq1\}\ y\  \{x\cdot y\leq\frac{2}{9}\})$$

Primero encontramos por separado las regiones del espacio $\Omega$ que satisfacen cada una de las desigualdades.
$$$$
1) Para $x+y\leq1$.
$$x+y=1 \qquad \Rightarrow \qquad y=1-x$$
Son todos los puntos que están por debajo de la recta $y=1-x$.
En la imagen, son todos los puntos en la región azul (y dentro del cuadrado $[0,1]$X$[0,1]$)

$$$$

2) Para $xy\leq\frac{2}{9}$.
$$xy=\frac{2}{9} \qquad \Rightarrow \qquad y=\frac{2}{9}x$$
Son todos los puntos que están por debajo de la hipérbola $ y=\frac{2}{9}x$. En la imagen, son todos los puntos en la región roja

.


El área donde se intersectan ambas regiones representa la probabilidad  $P(\{x+y\leq1\}\ y\  \{x\cdot y\leq\frac{2}{9}\})$

Ahora buscamos calcular dicho área.

Primero encontramos la intersección de ambas curvas.

\begin{align*}
\left\{ \begin{array}{l} y=1-x  \\ xy=\frac{2}{9}  \end{array} \right.
&&
\Rightarrow
&&
x(1-x)=\frac{2}{9}
&&
\Rightarrow
&&
x^2-x+\frac{2}{9} =0
\end{align*} 

\begin{align*}
\Rightarrow
&&
x&=\frac{1\pm\sqrt{1-4(1)(\frac{2}{9})}}{2}
\\
&& & \begin{array}{l} x_1=\frac{2}{3}  \\ x_1=\frac{1}{3}  \end{array} 
\end{align*} 



El área está determinada por la siguiente integral


\begin{align*}
Area &= \int_0^{1/3} (1-x)\ dx + \int_{1/3}^{2/3} \frac{2}{9}x\ dx + \int_{2/3}^1 (1-x)\ dx\\
&=(x-\frac{x^2}{2})\Bigg| _0^{1/3}    +   \frac{2}{9}ln(x)\Bigg|_{1/3}^{2/3}   +  (x-\frac{x^2}{2})\Bigg|_{2/3}^1\\
&\approx 0.487365
\end{align*}

$\therefore P(\{x+y\leq1\}\ y\  \{x\cdot y\leq\frac{2}{9}\})\approx 0.487365$

**PASO 7**. Obtenga la gráfica de estabilización que indica que el tamaño de muestra utilizado es suficiente para
garantizar la convergencia del resultado.

```python
import numpy as np
import random
from scipy.stats import norm
import math
import matplotlib.pyplot as plt
```

```python
def experimentos(n): #Definimos una función que realizara 
  acierto = 0 #Contador que ira sumando las veces que caiga acierto en N cantidad de veces.
  area=[] #Guardara las aproximaciones de la probabilidad en la iteración i.
  ensayos=[]
  for i in range(1,n+1):
      x = random.random()
      y = random.random()
      if x+y <= 1 and x*y<=(2/9) :
          acierto = acierto + 1
      ensayos.append(i) #Se va guardando en una lista la iteración i.
      area.append(acierto/i) #Se va guardadndo la aproximación a la probabilidad en la iteración i.
  return area #Regresamos una lista de las aproximaciones a la probabilidad de todas las iteraciones. 
```

```python
exp = experimentos(1500) #Hara el experimento antes descrito 1500 veces, es decir, nos dara 
#1500 aproximaciones a la probabilidad para ver cuando se logra estabilizar.
```

```python
np.mean(exp) #Saca la media de los experimentos, eso nos dará una idea de la aproximación al valor de la probabilidad.
```

```python
plt.plot(exp) #Grafica la lista de aproximaciones de acuerdo a la cantidad de experimentos realizados.
plt.xlabel('iteraciones')
plt.ylabel('Aproximación en la iteración i')
plt.title('Aproximación a la probabilidad')

# Agrega una línea vertical en la iteración 750 que de manera visual es donde comienza el estado estable.
plt.axvline(x=750, color='red', linestyle='--', label='Estado Estable en 750')

plt.legend()  # Muestra la leyenda si es necesario
plt.show()
```

**PASO 8**. Replique el modelo.

```python
def valor(repeticiones): #Esta función hara la simulación anterior muchas veces, es decir, hará "repeticiones" cantidad de veces
    Area = []
    for i in range(repeticiones):
        Area.append(np.mean(experimentos(750))) #Calcula la media de 750 experiementos unas N (repeticiones) cantidad de veces.
    return Area #Nos regresa N (repeticiones) medias de 750 experimentos cada una.
```

```python
N=100
simulacion = valor(N)
```

```python
np.mean(simulacion) #Saca la media de las N simulaciones realizadas, eso nos dará una mejor de la aproximación al valor de la probabilidad.
```

```python
plt.hist(simulacion) #Grafica las frecuencias de la aproximación al valor de la probabilidad de las N simualciones con 750 experiementos cada una.
plt.xlabel('iteraciones')
plt.ylabel('Frecuencia')
plt.title('Aproximación a la probabilidad')
plt.show()
```

**PASO 9**. Calcule la probabilidad y su desviación estándar.

```python
media = np.mean(simulacion) #Calcula la media de esas N simulaciones
var = np.var(simulacion) #Calcula la varianza de esas N simulaciones
desv = np.std(simulacion) #Calcula la desvaicion de esas N simulaciones
```

**PASO 10**. Halle el intervalo de confianza de la probabilidad con un nivel de confianza del $95 \%$.
\begin{equation*}
\overline{x}- Z_{\frac{\alpha}{2}}  \cdot \frac{S}{\sqrt{N}} < \mu < \overline{x}+ Z_{\frac{\alpha}{2}}  \cdot \frac{S}{\sqrt{N}}
\end{equation*}

```python
from scipy.stats import norm #Importamos la libreria para poder obtener el valor de Z.
alpha = 1 - 0.95 #Alpha es igual a 1 - la confianza del intervalo
z_alpha_2 = norm.ppf(1-alpha / 2) #Mandamos a pedir el valor de Z para \alpha/2, sabemos que si es el 95% de confienza, Z esta cercano al 1.96.
l_inferior = media - z_alpha_2 * (desv / math.sqrt(N)) #Sacamos limite inferior por la formula
l_superior = media + z_alpha_2 * (desv / math.sqrt(N)) #Sacamos limite superior por la formula
print("Entonces: ", l_inferior, "< \u03BC <", l_superior) #El intervalo en donde se encuentra el valor real a una confienza del 95%
```
